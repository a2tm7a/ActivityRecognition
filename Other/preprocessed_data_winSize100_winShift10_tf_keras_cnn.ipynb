```python
import h5py
import numpy as np
np.random.seed(1337)
from keras.optimizers import SGD
import logging
from keras.models import Sequential
from keras.layers import Dense, Dropout, Activation, Flatten
from keras.optimizers import SGD, Adam, RMSprop
from keras.utils import np_utils
from keras.regularizers import l2
from keras import backend as K
from keras.layers import Convolution2D, MaxPooling2D

```

```python

h5f = h5py.File('../preprocessed_data_winSize100_winShift10.h5','r')
X_train = h5f['training_data'][:]
y_train = h5f['training_output'][:]
X_test = h5f['testing_data'][:]
y_test = h5f['testing_output'][:]

h5f = h5py.File('../preprocessed_data_winSize100_winShift10_mean_sd.h5','r')
mean = h5f['mean'][:]
sd = h5f['sd'][:]
h5f.close()

print(X_train.shape,y_train.shape,X_test.shape,y_test.shape)
```

```python
print("Concatenating training")
training = np.concatenate([X_train, y_train.reshape(y_train.shape[0],1)], axis=1)
print("Deleting training")
del X_train
del y_train
print("Shuffling training")
np.random.shuffle(training)

print("Concatenating testing")
testing = np.concatenate([X_test, y_test.reshape(y_test.shape[0],1)], axis=1)
print("Deleting testing")
del X_test
del y_test
print("Shuffling Testing")
np.random.shuffle(testing)

print("Getting back train data and output")
X_train=training[:,:-1]
y_train=training[:,-1]
print("Deleting training")
del training

print("Getting back test data and output")
X_test=testing[:,:-1]
y_test=testing[:,-1]
print("Deleting testing")
del testing

print(X_train.shape,y_train.shape,X_test.shape,y_test.shape)
```

```python
X_train = X_train.astype('float32')
X_test = X_test.astype('float32')

X_test=(X_test-mean)/sd
X_train=(X_train-mean)/sd

print(X_train.shape,y_train.shape)
```

```python
nb_rows = 100
nb_cols = 3

if K.image_dim_ordering() == 'th':
    X_train = X_train.reshape(X_train.shape[0], 1, nb_rows, nb_cols)
    X_test = X_test.reshape(X_test.shape[0], 1, nb_rows, nb_cols)
    input_shape = (1, nb_rows, nb_cols)
else:
    X_train = X_train.reshape(X_train.shape[0], nb_rows, nb_cols, 1)
    X_test = X_test.reshape(X_test.shape[0], nb_rows, nb_cols, 1)
    input_shape = (nb_rows, nb_cols, 1)
```

```python
nb_classes = 7
y_train = np_utils.to_categorical(y_train, nb_classes)
y_test = np_utils.to_categorical(y_test, nb_classes)
print(y_train.shape)
```

```python

print(X_train.shape)

batch_size = 512

nb_epoch = 10
nb_neg_cycles = 3

logging.basicConfig(filename= "./WSDM_Keras_batch_size"+"("+str(batch_size)+")"+".log", level=logging.INFO, format='%(message)s')
logging.info("Parameters")
```

```python

# number of convolutional filters to use
nb_filters = 180
# size of pooling area for max pooling
pool_size = (20, 1)
# convolution kernel size
kernel_size = (60, 3)

print(X_train.shape)

```

```python
model = Sequential()

model.add(Convolution2D(nb_filters, kernel_size[0], kernel_size[1],
                        border_mode='valid',
                        input_shape=input_shape))
model.add(Activation('relu'))
#model.add(MaxPooling2D(pool_size=pool_size,strides=(2,1)))

model.add(Flatten())
# model.add(Dense(128))
# model.add(Activation('relu'))
# model.add(Dropout(0.5))
model.add(Dense(nb_classes))
model.add(Activation('softmax'))
model.summary()
```

```python

epoch = 1

final_acc=0.0

lr = 0.002

weights=[]
first_run=True
flag=True
iter_count = 0
count_neg_iter=0


#The flag to keep the loop running
run_flag=True
weights=[]

#Check if it is the first iteration
first_iter=True
```

```python
while run_flag:

    #Give ramdom weights in first iteration and previous weights to other
    if first_iter:
        first_iter=False
    else:
        model.set_weights(np.asarray(weights))


    logging.info("\n\nIteration:"+str(iter_count))
    print ("Learning Rate : ",lr)
    logging.info("Learning Rate: "+str(lr))
    
    sgd = SGD(lr=lr)
    
    model.compile(loss='categorical_crossentropy',
                  optimizer=sgd,
                  metrics=['accuracy'])

    model.fit(X_train, y_train, batch_size=batch_size, nb_epoch=1,
              verbose=1, validation_data=(X_test, y_test))
    score = model.evaluate(X_test, y_test, verbose=0)
    
    
    print('Test score:', score[0])
    print('Test accuracy:', score[1])
    
    current_acc = score[1]

    logging.info(current_acc)

    if(current_acc - final_acc > 0.0001):        
        iter_count = iter_count + 1

        #Update the weights if the accuracy is greater than .001
        weights=model.get_weights()
        print ("Updating the weights")
        logging.info("Updating the weights")
        #Updating the final accuracy
        final_acc=current_acc
        #Setting the count to 0 again so that the loop doesn't stop before reducing the learning rate n times consecutivly
        count_neg_iter = 0

    else:
        #If the difference is not greater than 0.005 reduce the learning rate
        lr=lr/2.0
        print ("Reducing the learning rate by half")
        logging.info("Reducing the learning rate by half")
        count_neg_iter = count_neg_iter + 1
        
        #If the learning rate is reduced consecutively for nb_neg_cycles times then the loop should stop
        if(count_neg_iter>nb_neg_cycles):
            run_flag=False
            model.set_weights(np.asarray(weights))
```

```python
print('Final accuracy:', final_acc)
print(iter_count)
logging.info(final_acc)
```

```python
y_predict = model.predict_classes(X_test)
y_actual=np.argmax(y_test,axis=1)
correct=0
total=0
for i in range(0,y_predict.shape[0]):
    if(y_predict[i]==y_actual[i]):
        correct+=1
    total+=1
print(float(correct/total))
print(y_predict)
target_names = ['class 0(walking)', 'class 1(standing)', 'class 2(jogging)', 'class 3(sitting)', 'class 4(biking)', 'class 5(upstairs)', 'class 6(downstairs)']
```

```python
print(np.argmax(y_test,axis=1).shape)

from sklearn.metrics import classification_report,confusion_matrix
print(classification_report(np.argmax(y_test,axis=1), y_predict,target_names=target_names))
print(confusion_matrix(np.argmax(y_test,axis=1), y_predict))
```
